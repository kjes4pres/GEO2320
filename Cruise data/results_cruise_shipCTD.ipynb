```python
def plot_hydrographic(station_name, df1, df2):

    salinity1 = df1["Salinity, Practical [PSU]"]
    temperature1 = df1["Temperature [ITS-90, deg C]"]
    density1 = df1["Density [sigma-t, kg/m^3]"]
    depth1 = df1["Pressure, Digiquartz [db]"]

    salinity2 = df2["Salinity, Practical [PSU]"]
    temperature2 = df2["Temperature [ITS-90, deg C]"]
    density2 = df2["Density [sigma-t, kg/m^3]"]
    depth2 = df2["Pressure, Digiquartz [db]"]    

    fig, ax = plt.subplots(nrows=1, ncols=3)
    fig.suptitle(station_name)

    ax[0].plot(salinity1, depth1, label = "Station 1")
    ax[1].plot(temperature1, depth1, label = "Station 1")
    ax[2].plot(density1, depth1, label = "Station 1")

    ax[0].plot(salinity2, depth2, label = "Station 2")
    ax[1].plot(temperature2, depth2, label = "Station 2")
    ax[2].plot(density2, depth2, label = "Station 2")

    ax[0].legend()
    ax[1].legend()
    ax[2].legend()

    ax[0].set_title("Salinity")
    ax[1].set_title("Temperature")
    ax[2].set_title("Density")

    ax[0].set_xlabel("Salinity")
    ax[0].set_ylabel("Depth [m]")
    ax[0].invert_yaxis()

    ax[1].set_xlabel("Temperature (Celsius)")
    ax[1].set_ylabel("Depth [m]")
    ax[1].invert_yaxis()

    ax[2].set_xlabel("Density [kgm^-3]")
    ax[2].set_ylabel("Depth [m]")
    ax[2].invert_yaxis()

    for ax in ax.flat:
        ax.grid(True)

    fig.text(0.5, 0.92, "Results from ship CTD", ha='center', fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
plot_hydrographic(f"Station 1 & 2, leg 1, UTC: 07:17, 08:11", df_0717, df_0811)
```

```python
def plot_hydrographic(station_name, df1, df2, df3):

    salinity1 = df1["Salinity, Practical [PSU]"]
    temperature1 = df1["Temperature [ITS-90, deg C]"]
    density1 = df1["Density [sigma-t, kg/m^3]"]
    depth1 = df1["Pressure, Digiquartz [db]"]

    salinity2 = df2["Salinity, Practical [PSU]"]
    temperature2 = df2["Temperature [ITS-90, deg C]"]
    density2 = df2["Density [sigma-t, kg/m^3]"]
    depth2 = df2["Pressure, Digiquartz [db]"]   

    salinity3 = df3["Salinity, Practical [PSU]"]
    temperature3 = df3["Temperature [ITS-90, deg C]"]
    density3 = df3["Density [sigma-t, kg/m^3]"]
    depth3 = df3["Pressure, Digiquartz [db]"]  

    fig, ax = plt.subplots(nrows=1, ncols=3)
    fig.suptitle(station_name)

    ax[0].plot(salinity1, depth1, label = "Station 1")
    ax[1].plot(temperature1, depth1, label = "Station 1")
    ax[2].plot(density1, depth1, label = "Station 1")

    ax[0].plot(salinity2, depth2, label = "Station 2")
    ax[1].plot(temperature2, depth2, label = "Station 2")
    ax[2].plot(density2, depth2, label = "Station 2")

    ax[0].plot(salinity3, depth3, label = "Station 3")
    ax[1].plot(temperature3, depth3, label = "Station 3")
    ax[2].plot(density3, depth3, label = "Station 3")

    ax[0].legend()
    ax[1].legend()
    ax[2].legend()

    ax[0].set_title("Salinity")
    ax[1].set_title("Temperature")
    ax[2].set_title("Density")

    ax[0].set_xlabel("Salinity")
    ax[0].set_ylabel("Depth [m]")
    ax[0].invert_yaxis()

    ax[1].set_xlabel("Temperature (Celsius)")
    ax[1].set_ylabel("Depth [m]")
    ax[1].invert_yaxis()

    ax[2].set_xlabel("Density [kgm^-3]")
    ax[2].set_ylabel("Depth [m]")
    ax[2].invert_yaxis()

    for ax in ax.flat:
        ax.grid(True)

    fig.text(0.5, 0.92, "Results from ship CTD", ha='center', fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
plot_hydrographic(f"Station 1, 2 and 3, leg 2, UTC: 11:12, 12:05, 12:46", df_1112, df_1205, df_1246)
```
