```python
def plot_hydrographic(station_name, dataframe):

    salinity = dataframe["Salinity, Practical [PSU]"]
    temperature = dataframe["Temperature [ITS-90, deg C]"]
    density = dataframe["Density [sigma-t, kg/m^3]"]
    depth = dataframe["Pressure, Digiquartz [db]"]

    fig, ax = plt.subplots(nrows=1, ncols=3)
    fig.suptitle(station_name)

    ax[0].plot(salinity, depth)
    ax[1].plot(temperature, depth, color = "r")
    ax[2].plot(density, depth, color = "g")

    ax[0].set_title("Salinity")
    ax[1].set_title("Temperature")
    ax[2].set_title("Density")

    ax[0].set_xlabel("Salinity")
    ax[0].set_ylabel("Depth [m]")
    ax[0].invert_yaxis()

    ax[1].set_xlabel("Temperature (Celsius)")
    ax[1].set_ylabel("Depth [m]")
    ax[1].invert_yaxis()

    ax[2].set_xlabel("Density [kgm^-3]")
    ax[2].set_ylabel("Depth [m]")
    ax[2].invert_yaxis()

    for ax in ax.flat:
        ax.grid(True)

    fig.text(0.5, 0.92, "Results from ship CTD", ha='center', fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
plot_hydrographic(f"Station 1, leg 1, UTC: 07:17, latitude: {lat_0717}, longitude: {lon_0717}", df_0717)
```
