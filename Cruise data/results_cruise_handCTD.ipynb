# Results from hand held CTDs, at six stations.

```python
import gsw
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def make_dataframe(datapath):
    df_metadata = pd.read_csv(datapath, header=None, nrows=28)
    df_data = pd.read_csv(datapath, header=None,skiprows=29)
    df_data.columns = ["Pressure (Decibar)", "Depth (Meter)", "Temperature (Celsius)", "Conductivity (MicroSiemens per Centimeter)", "Specific conductance (MicroSiemens per Centimeter)", "Salinity (Practical Salinity Scale)", "Sound velocity (Meters per Second)", "Density (Kilograms per Cubic Meter)"
]

    lat = df_metadata.iloc[9, 1]
    lon = df_metadata.iloc[10, 1]
    
    return df_metadata, df_data, lon, lat
```

```python
def plot_hydrographic(station_name, df1, df2):

    salinity1 = df1["Salinity (Practical Salinity Scale)"]
    temperature1 = df1["Temperature (Celsius)"]
    density1 = df1["Density (Kilograms per Cubic Meter)"]
    depth1 = df1["Pressure (Decibar)"]

    salinity2 = df2["Salinity (Practical Salinity Scale)"]
    temperature2 = df2["Temperature (Celsius)"]
    density2 = df2["Density (Kilograms per Cubic Meter)"]
    depth2 = df2["Pressure (Decibar)"]

    fig, ax = plt.subplots(nrows=1, ncols=3)
    fig.suptitle(station_name)

    ax[0].plot(salinity1, depth1, label = "Station 1")
    ax[1].plot(temperature1, depth1, label = "Station 1")
    ax[2].plot(density1, depth1, label = "Station 1")
    ax[0].plot(salinity2, depth2, label = "Station 2")
    ax[0].legend()
    ax[1].plot(temperature2, depth2, label = "Station 2")
    ax[1].legend()
    ax[2].plot(density2, depth2, label = "Station 2")
    ax[2].legend()

    ax[0].set_title("Salinity")
    ax[1].set_title("Temperature")
    ax[2].set_title("Density")

    ax[0].set_xlabel("Salinity")
    ax[0].set_ylabel("Depth [m]")
    ax[0].invert_yaxis()

    ax[1].set_xlabel("Temperature (Celsius)")
    ax[1].set_ylabel("Depth [m]")
    ax[1].invert_yaxis()

    ax[2].set_xlabel("Density [kgm^-3]")
    ax[2].set_ylabel("Depth [m]")
    ax[2].invert_yaxis()

    for ax in ax.flat:
        ax.grid(True)

    fig.text(0.5, 0.92, "Results from hand held CTD", ha='center', fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
meta_0717, data_0717, lon_0717, lat_0717 = make_dataframe("C:\\Users\\stang\\GEO2320\\Cruise data\\handheld_ctd\\CC2240001_20230928_071740.csv")
meta_0811, data_0811, lon_0811, lat_0811 = make_dataframe("C:\\Users\\stang\\GEO2320\\Cruise data\\handheld_ctd\\CC2240001_20230928_081103.csv")
plot_hydrographic(station_name=f"Station 1 and 2, leg 1, UTC: 07:17 and UTC 08:11",df1= data_0717, df2 =data_0811)
```
